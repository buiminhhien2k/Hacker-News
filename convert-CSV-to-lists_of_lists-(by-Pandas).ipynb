```python
import pandas as pd
hn_panda=pd.read_csv("HN_posts_year_to_Sep_26_2016.csv",header=None)
header=hn_panda.iloc[0]
hn_header=list(header)
```

```python
hn_panda=hn_panda[1:]
hn_panda=hn_panda.rename(columns=header)
```

```python
hn_panda=hn_panda.astype(str)# converting Data Frame into 'object', so that the filter below run smoothly without any omission
hn_panda=hn_panda.loc[hn_panda['num_comments']!='0']
size=hn_panda.shape
hn_panda.index=range(size[0]) #remaking a new row-index for Date Frame base on the shape of the Data Frame
hn=[]
for i in range(0,size[0]):
    i=list(hn_panda.loc[i])
    hn.append(i)
```

```python
print(hn_header)
for each_row in hn[0:3]:
    print("\n")
    print(each_row)
print("\n",len(hn))
```
